```python
from src.utils.agent_models import AgentResponseModel, Message, Role
from src.sql_agent.data_query import SQLQueryAgent
from src.sql_agent.api import sql_agent_conversation
from src.utils.global_config import GlobalSettings
from src.utils.instance import Instance

active_instance : Instance = GlobalSettings.instance_by_id(GlobalSettings().active_instance_id)
msg = Message(role="user", content=user_query)
conversation = AgentResponseModel()
conversation.add_message(msg)

conversation = sql_agent_conversation(client_id=active_instance.ClientUID, project_id=active_instance.ProjectUID, model=conversation)
display(conversation.model_dump())
```

```python
import json
conversation.follow_up_questions[0].answers = submodule

conversation = sql_agent_conversation(client_id=active_instance.ClientUID, project_id=active_instance.ProjectUID, model=conversation)
display(conversation.model_dump())
```

```python
from src.utils.operators import StringOperators, NumericOperators, NumericValue
from src.utils.agent_models import FollowUpQuestion

vendorcountry = 'US'
vendorcountryname = 'UNITED STATES'
for q in conversation.follow_up_questions:
    q : FollowUpQuestion
    if q.column_name.lower() == 'vendorcountry':
        q.answers = [vendorcountry]
        q.answer_operator = StringOperators.EQUALS
    if q.column_name.lower() == 'vendorcountryname':
        q.answers = [vendorcountryname]
        q.answer_operator = StringOperators.EQUALS
    if q.column_name.lower() == 'fiscalyear':
        q.answers = NumericValue(values=[2020])
        q.answer_operator = NumericOperators.EQUALS

conversation = sql_agent_conversation(client_id=active_instance.ClientUID, project_id=active_instance.ProjectUID, model=conversation)
display(json.loads(conversation.model_dump_json()))
```
