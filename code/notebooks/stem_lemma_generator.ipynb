```python
import pandas as pd

file_path = r'C:\Users\SWECSAT056\Downloads\OCR Tests.xlsx'

# open excel file
df = pd.read_excel(file_path, sheet_name='Gambling')
df
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

# update the Stem column with stem of the word
def stem_lemma_generator(df, column_name, stem=True, lemma=True):
    if stem:
        stemmer = nltk.stem.PorterStemmer()
        df['Stem'] = df[column_name].apply(lambda x: stemmer.stem(str(x).lower()))
    if lemma:
        lemmatizer = WordNetLemmatizer()
        df['Lemma'] = df[column_name].apply(lambda x: lemmatizer.lemmatize(str(x).lower()))

    return df

# test
df = stem_lemma_generator(df, 'Keyword', stem=True, lemma=True)

df
```

```python
# write back to excel
df.drop_duplicates(inplace=True)
df.to_excel(r'C:\Users\SWECSAT056\Documents\KonaAI_ML\code\notebooks\stem_lemma.xlsx', index=False)
```
