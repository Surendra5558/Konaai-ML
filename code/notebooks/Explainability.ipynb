```python
from src.automl.explainer import PredictionExplainer, ExplainationOutput
from src.automl.UI.explainations import PredictionExplainerUI
from src.utils.conf import Setup
from src.utils.submodule import Submodule
from src.utils.db_mgmt import DBHandler
from src.automl.utils import config
from src.tools.dask_tools import validate_column

import matplotlib.pyplot as plt
import pandas as pd
import dask.dataframe as dd
import ipywidgets as widgets
from ipywidgets import interact
import numpy as np
from IPython.display import display

sub = Submodule('P2P', 'Invoices', Setup().active_instance_id)


explainerui = PredictionExplainerUI(sub)
explainerui._chose_sample_transactions()
print(explainerui.sample_transactions)

index_column = validate_column('spt_rowid', explainerui.sample_transactions)

def plot_bar_chart(data : dict):
    df = pd.DataFrame(data)
    # plot horizontal bar chart
    fig, ax = plt.subplots(figsize=(10, len(df)*0.3))

    # set size of the plot
    colors = ['red' if x > 0 else 'green' for x in df['contribution_percent'].values]

    ax.barh(df['name'], df['contribution_percent'], color=colors)
    ax.set_xlabel('Contribution Percent')
    ax.set_ylabel('Feature')
    ax.set_title('Feature Contribution Percent')
    plt.show()

    print(df)


@interact(transaction_id=explainerui.sample_transactions[index_column].values)
def explain(transaction_id : int):
    print('Transaction ID', transaction_id)
    explainer = PredictionExplainer(submodule_obj=sub, transaction_id=transaction_id)
    if explaination := explainer.explain():
        for k, v in explaination.model_dump().items():
            if k == "features":
                plot_bar_chart(v)
                continue
            print(k, ":", v)
```
