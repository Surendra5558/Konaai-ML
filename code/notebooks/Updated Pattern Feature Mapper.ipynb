```python
from pathlib import Path
import os
import json

feature_file_path = os.path.join(Path.cwd().parent, r'src\assets\ml_feature_mapping.json')
feature_mappings = json.load(open(feature_file_path))

def get_feature_details(pattern_id):
    # each key in feature_mappings is a dict with name and description keys. filter for the feature_name
    feature_details = list(filter(lambda x: x['PatternID'] == pattern_id, feature_mappings.values()))
    return feature_details[0] if feature_details else None

# sourcery skip: raise-specific-error
TABLE_NAME = "[Model].[Patterns]"
test_prefix = 'FMP2P'

test_start_index = 1001
test_end_index = 1080

feature_mapper = r"code\src\assets\ml_feature_mapping.json"

for i in range(test_start_index, test_end_index+1):
    pattern_id = f"{test_prefix}{i:03}"
    print(f"-- {pattern_id}")

    feature_details = get_feature_details(pattern_id)
    if not feature_details:
        raise Exception(f'Feature details not found for pattern_id: {pattern_id}')


    model_patterns = Model_Patterns_Template.copy()
    for key, value in model_patterns.items():
        if value is None:
            model_patterns[key] = (
                feature_details[key] if key in feature_details else None
            )

    if none_keys := [
        f"{key}:{value}"
        for key, value in model_patterns.items()
        if value is None
    ]:
        raise Exception(f'None keys: {none_keys}')


    # prepare an insert statement from model_patterns
    column_statement = ', '.join([f"[{pattern}]" for pattern in model_patterns.keys()])
    insert_statement = f"INSERT {TABLE_NAME} ({column_statement})"
    values_statement = ', '.join(
        [f"'{value}'" 
        if isinstance(value, str) and value != 'NULL' and value != 'GETUTCDATE()' else str(value) 
        for value in model_patterns.values()]
        )
    insert_statement = f"{insert_statement} VALUES ({values_statement})"
    print(insert_statement)
```
